```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data["timestamp"] = le.fit_transform(data["timestamp"])
data["diabetes_in_family"] = le.fit_transform(data["diabetes_in_family"])
data["late_sleeping"] = le.fit_transform(data["late_sleeping"])
data["late_rising"] = le.fit_transform(data["late_rising"])
data["regular_exercise"] = le.fit_transform(data["regular_exercise"])
data["Sleeping_after_eat"] = le.fit_transform(data["Sleeping_after_eat"])
data["drug_addicted"] = le.fit_transform(data["drug_addicted"])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
average = 'macro'

from sklearn import metrics
file_csv = [['Name', 'Accuracy', 'F1S', 'Rs', 'PS', 'FBS', 'HL', 'JS', 'MC', 'AUC', 'BAC', 'CKS']]

def performance(name, y_test, y_pred, average):
    # Accuracy classification score
    accuracy = metrics.accuracy_score(y_test, y_pred)
    print('accuracy = ',accuracy)

    # https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html#sklearn.metrics.f1_score
    # Compute the F1 score, also known as balanced F-score or F-measure.
    # if binary average = 'binary'
    f1s = metrics.f1_score(y_test, y_pred, average=average)
    print('F1S = ',f1s)


    # https://scikit-learn.org/stable/modules/generated/sklearn.metrics.precision_score.html#sklearn.metrics.precision_score
    # Compute the recall score
    # if binary average = 'binary'
    rs = metrics.recall_score(y_test, y_pred, average=average)
    print('RS = ',rs)


    # https://scikit-learn.org/stable/modules/generated/sklearn.metrics.precision_score.html#sklearn.metrics.precision_score
    # Compute the precision score
    # if binary average = 'binary'
    ps = metrics.precision_score(y_test, y_pred, average=average)
    print('PS = ',ps)
```

```python
from sklearn.ensemble import AdaBoostClassifier
clf = AdaBoostClassifier(n_estimators=50,  algorithm='SAMME.R')
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
performance('AdC',y_test, y_pred, average)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100,  criterion='gini')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
performance('RFC',y_test, y_pred, average)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5,  algorithm='auto', metric='minkowski')
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
performance('KNC', y_test, y_pred, average)
```

```python
from sklearn.naive_bayes import BernoulliNB
clf = BernoulliNB( binarize=0.0)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
performance('BNB', y_test, y_pred, average)
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(random_state=1, max_iter=300)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
performance('MLPC', y_test, y_pred, average)
```
